**Problema 1**

---



```python
import numpy as np
import matplotlib.pyplot as plt

# Problema 1A
def func(x):
    f = np.exp(-x)-x
    return f

x = np.linspace(-1,1,100)
plt.plot(x,func(x),'k-')

y = np.zeros(100)

plt.plot(x,y, 'r')

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("HW 1 Problema 1: f(x) vs. x")
plt.show()

# Problema 1D
def f_iterate(x):
    xnew = np.exp(-x)
    return xnew

xold = 0.0      # Chute inicial
xnew = f_iterate(xold)
erro = np.abs(xnew-xold)
tol = 0.001

k = 0           # Contador de iterações

while erro>tol:
    xold = xnew
    xnew = f_iterate(xold)
    erro = np.abs(xnew-xold)
    k = k + 1
    if k==100:
        print('Não convergiu em 1000 passos')
        break

print(f"O raiz de f(x) é x = {xnew}")
print(f"O código convergiu em k = {k} iterações")
```

**Problema 2**

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Problema 2A
def conc(t):
    c = 80*np.exp(-2*t)+20*np.exp(-0.1*t)
    return c

t = np.linspace(1,15,100)
plt.plot(t,conc(t),'k-')

y = 10*np.ones(100)
plt.plot(t,y, 'r')

plt.xlabel("tempo (min)")
plt.ylabel("Concentração (mol/L)")
plt.title("HW 1 Problema 2: C(t) vs. t")
plt.show()

# Problema 2B
def c_iterate(t):
    tnew = 80*np.exp(-2*t)+20*np.exp(-0.1*t)-10+t
    return tnew

told = 0.0      # Chute inicial
tnew = c_iterate(told)
erro = np.abs(tnew-told)
tol = 0.001

k = 0           # Contador de iterações

while erro>tol:
    told = tnew
    tnew = c_iterate(told)
    erro = np.abs(tnew-told)
    k = k + 1
    if k==100:
        print('Não convergiu em 1000 passos')
        break

print(f"O raiz de C(t) é t = {tnew} min")
print(f"O código convergiu em k = {k} iterações")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Problema 3
def vel(t):
    v = np.zeros(len(t))

    for i in range(len(t)):
        if 0<=t[i]<10:
            v[i] = 11*t[i]**2-5*t[i]
        elif 10<=t[i]<20:
            v[i] = 1100-5*t[i]
        elif 20<=t[i]<30:
            v[i] = 5*t[i]+2*(t[i]-20)**2
        else:
            v[i] = 0
    return v

t = np.linspace(-10,60,500)

plt.plot(t,vel(t),'k-')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.title('HW 1 Problema 3: v(t) vs. t')
plt.show()
```
