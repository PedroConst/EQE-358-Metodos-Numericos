```python
import numpy as np
import matplotlib.pyplot as plt
```

## **Problema 1:** Cálculo do determinante

```python
# Problema 1: Cálculo do determinante
matrizes = [
    np.array([[1, 2], [2, 4]]),
    np.array([[5, -3, 2], [1, 0, 2], [2, -1, 3]]),
    np.array([[1, 0, 3, -2], [6, 1, -1, 3], [2, 0, 1, 1], [4, 3, 2, 5]])
]

for i, A in enumerate(matrizes, 1):
    det = np.linalg.det(A)
    print(f"Determinante da matriz {i}: {det}")
    if np.isclose(det, 0):
        print("A matriz é singular.")
```

## **Problema 2:** Cálculo da inversa e verificação de ortogonalidade

```python
# Problema 2: Cálculo da inversa e verificação de ortogonalidade
matrizes_inv = [
    np.array([[1, 1], [1, -1]]),
    np.array([[1, 2], [3, 4]]),
    np.array([[1, 0], [0, -1]]),
    np.array([[2, 1, 3], [0, 1, 1], [0, 0, 1]])
]

for i, A in enumerate(matrizes_inv, 1):
    det = np.linalg.det(A)
    if np.isclose(det, 0):
        print(f"Matriz {i} não tem inversa (det=0).")
    else:
        A_inv = np.linalg.inv(A)
        print(f"Inversa da matriz {i}:\n{A_inv}")
        if np.allclose(A.T @ A, np.eye(A.shape[0])):
            print("A matriz é ortogonal.")
```

## **Problema 3:** Resolução de sistemas lineares

```python
# Problema 3: Resolução de sistemas lineares
sistemas = [
    (np.array([[1, 2], [-1, 4]]), np.array([2, 1])),
    (np.array([[1, 1], [3, -2]]), np.array([2, 5])),
    (np.array([[1, 1, -1], [1, 1, 1], [1, -1, 0]]), np.array([1, 2, 3]))
]

for i, (A, b) in enumerate(sistemas, 1):
    solucao = np.linalg.solve(A, b)
    print(f"Solução do sistema {i}: {solucao}")
```

## **Problema 4:** Função da velocidade do foguete

```python
# Problema 4: Função da velocidade do foguete
def velocidade(t):

    v= np.zeros(len(t))

    for i in range(len(t)):
        if 0 <= t[i] < 10:
            v[i] = 11*t[i]**2 - 5*t[i]
        elif 10 <= t[i] < 20:
            v[i] = 1100 - 5*t[i]
        elif 20 <= t[i] < 30:
            v[i] = 5*t[i] + 2*(t[i]-20)**2
        else:
            v[i] = 0
    return v

t_vals = np.linspace(-10, 60, 1000)
v_vals = velocidade(t_vals)

plt.plot(t_vals, v_vals, label='v(t)')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.title('Velocidade do foguete vs Tempo')
plt.legend()
plt.grid()
plt.show()
```

## **Problema 5:** Volume do líquido no cilindro



```python
# Problema 5: Volume do líquido no cilindro
def volume_cilindro(h):

    r =2    # raio (m)
    L = 5   # comprimento (m)
    Vol = np.zeros(len(h))

    for i in range(len(h)):
        if h[i] > 2*r or h[i] < 0:
            Vol[i] = 0
        Vol[i] = (r**2 * np.acos((r-h[i])/r) - (r-h[i]) * np.sqrt(2*r*h[i] - h[i]**2)) * L

    return Vol

h_vals = np.linspace(0, 4, 100)
V_vals = volume_cilindro(h_vals)

plt.plot(h_vals, V_vals, label='V(h)')
plt.xlabel('Profundidade (m)')
plt.ylabel('Volume (m³)')
plt.title('Volume do líquido vs Profundidade')
plt.legend()
plt.grid()
plt.show()
```
